```python
AppLaunched["Hours"]=AppLaunched["Minute_Of_Day"]//60
AppLaunched["Minutes"]=AppLaunched["Minute_Of_Day"]%60
AppLaunched["Hours"]=AppLaunched["Hours"].astype(str)
AppLaunched["Minutes"]=AppLaunched["Minutes"].astype(str)
AppLaunched["Second"]=AppLaunched["Second"].astype(str)
AppLaunched['time'] = AppLaunched[['Hours', 'Minutes']].apply(lambda x: ':'.join(x), axis=1)
AppLaunched['time_sec']=AppLaunched[['time', 'Second']].apply(lambda x: ':'.join(x), axis=1)
AppLaunched["str_date"]=AppLaunched[['Date']].applymap(str).applymap(lambda s: "{}-{}-{}".format(s[6:],s[4:6], s[0:4]))
AppLaunched['date_time']=AppLaunched[['str_date', 'time_sec']].apply(lambda x: ' '.join(x), axis=1)
AppLaunched["Timestamp"]=pd.to_datetime(AppLaunched.date_time, format="%d-%m-%Y %H:%M:%S")
AppLaunched["Timestamp"] = AppLaunched["Timestamp"].dt.strftime('%d-%m-%Y %H:%M:%S')
columns = ['Date', 'Minute_Of_Day','Second','Hours','Minutes','time','time_sec','str_date','date_time']
AppLaunched.drop(columns, inplace=True, axis=1)
```

```python
sortedMatrix = np.argsort(corr_matrix,axis=1)
dictRecommend = {}
for user in sortedMatrix.columns:
    indexes = sortedMatrix[user][-5:]
    user_list = [sortedMatrix.columns[ind] for ind in indexes]
    dictRecommend[user] = user_list
```

```python
#past data
uniqueVideo = {}
for k,v in userVdoCount.items():
    key = k[0]
    value = k[1]
    if k[0] in uniqueVideo:
        uniqueVideo[k[0]].append(k[1])
    else:
        uniqueVideo[k[0]] = [k[1]]
```

```python
#we found  recommended videos wrt user
finalRecDict = {}
for user in dictRecommend:
    vList = []
    for userrec in dictRecommend[user]:
        vList += uniqueVideo[userrec]
    finalRecDict[user] = vList
    
```

```python
correct = 0
incorrect = 0
for user in uniqueVideo:
    s = set(uniqueVideo[user]).intersection(set(finalRecDict[user]))
    correct += len(s) 
    incorrect += len(set(uniqueVideo[user])-set(finalRecDict[user]))
```
