```python
import pandas as pd
```

```python
VideoDetails= pd.read_csv("C:/Users/VIDYA/Desktop/Capstone/Data/VideoDetails.csv")
```

```python
VideoDetails.head()
```

```python
VideoDetails["Hours"]=VideoDetails["Minute_Of_Day"]//60
VideoDetails["Minutes"]=VideoDetails["Minute_Of_Day"]%60
VideoDetails["Hours"]=VideoDetails["Hours"].astype(str)
VideoDetails["Minutes"]=VideoDetails["Minutes"].astype(str)
VideoDetails["Second"]=VideoDetails["Second"].astype(str)
VideoDetails['time'] = VideoDetails[['Hours', 'Minutes']].apply(lambda x: ':'.join(x), axis=1)
VideoDetails['time_sec']=VideoDetails[['time', 'Second']].apply(lambda x: ':'.join(x), axis=1)
VideoDetails["str_date"]=VideoDetails[['Date']].applymap(str).applymap(lambda s: "{}-{}-{}".format(s[6:],s[4:6], s[0:4]))
VideoDetails['date_time']=VideoDetails[['str_date', 'time_sec']].apply(lambda x: ' '.join(x), axis=1)
VideoDetails["Timestamp"]=pd.to_datetime(VideoDetails.date_time, format="%d-%m-%Y %H:%M:%S")
VideoDetails["Timestamp"] = VideoDetails["Timestamp"].dt.strftime('%d-%m-%Y %H:%M:%S')
columns = ['Date', 'Minute_Of_Day','Second','Hours','Minutes','time','time_sec','str_date','date_time']
VideoDetails.drop(columns, inplace=True, axis=1)
```

```python
VideoDetails.rename(columns={'State':'State_vdts','Genre':'Genre_vdts','Category':'Category_vdts','ProgramType':'ProgramType_vdts','Country':'Country_vdts','Device':'Device_vdts','OS':'OS_vdts','Timestamp':'Timestamp_vdts'}, inplace=True)
```
